```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.array([4.0,3.0,2.0])
ys = np.array([3.0,2.0,1.0])
n = len(xs)
```

```python
def loss_and_grads(w,b,xs,ys):
    preds = w*xs + b
    errs = preds - ys
    loss = 0.5 * np.mean(errs**2)
    dw = np.mean(errs * xs)   # derivative of 0.5*mean((pred-ys)^2) wrt w
    db = np.mean(errs)
    return loss, dw, db
```

```python
# init
w, b = 0.0, 0.0
lr = 0.1
```

```python
history = []
for it in range(1000):
    loss, dw, db = loss_and_grads(w,b,xs,ys)
    history.append((it, w, b, loss))
    w -= lr * dw
    b -= lr * db
```

```python
for it,wv,bv,l in history:
    print(f"iter {it}: w={wv:.9f}, b={bv:.9f}, loss={l:.9f}")
```

```python
# Plot final line and data
xline = np.linspace(1.5,4.5,200)
plt.scatter(xs, ys, label='data')
plt.plot(xline, w*xline + b, label=f'GD final w={w:.4f}, b={b:.4f}')
plt.legend()
plt.show()
```
